```python
# Definição das variáveis de entrada
Vi = 15            #  Digite a tensão de entrada (Vi)
Vo =2             #  Digite a tensão de saída (Vo)
#P0 = 10            #  Digite a potência máxima de saída (P0) P0=V0*I0
Io= 2
P0 = Vo * Io # Potência de saída (W)
fs = 60*10**3       #  Digite a frequência de comutação (fs)
dIL = 0.05   # Ripple de corrente no indutor (20%)
dVc = 0.01  # Ripple de tensão no capacitor (1%)

# Cálculo do ciclo de trabalho
D = Vo / Vi
print(f"O ciclo de trabalho é: {D:.4f}")

# Cálculo da corrente de saída
I0 = P0 / Vo
print(f"A corrente de saída é: {I0:.4f} A")

# Cálculo da resistência de carga
R0 = Vo**2 / P0
print(f"A resistência de carga é: {R0:.4f} Ω")

# Cálculo do indutor
L0 = ((Vi - Vo) * D / (fs * (dIL * I0))*10**6)

print(f"O indutor é igual a: {L0:.4f} uH")

# Cálculo da ondulação de corrente no indutor
dIL_calc = (Vi - Vo) * D / (fs * (L0*10**-6))
print(f"A ondulação de corrente no indutor é: {dIL_calc:.4f} A")

# Cálculo da capacitância
C0 = dIL_calc / (8 * fs  * (dVc*Vo))*10**6
print(f"A capacitância é igual a: {C0:.4f} uF")

# Cálculo do esforço na chave
Ismd = D * I0
Isrms = (D**0.5) * I0
Ismax = I0 + dIL / 2
Vsmax = Vi

print(f"Esforço na chave:")
print(f"  Valor médio da corrente: {Ismd:.4f} A")
print(f"  Valor eficaz da corrente: {Isrms:.4f} A")
print(f"  Valor máximo de corrente: {Ismax:.4f} A")
print(f"  Valor máximo de tensão: {Vsmax:.4f} V")

# Cálculo do diodo
Idmd = (1 - D) * I0
Idrms = ((1 - D) ** 0.5) * I0
Idmax = I0 + dIL / 2

print(f"Cálculos do diodo:")
print(f"  Valor médio da corrente: {Idmd:.4f} A")
print(f"  Valor eficaz da corrente: {Idrms:.4f} A")
print(f"  Valor máximo de corrente: {Idmax:.4f} A")
print(f"  Valor máximo de tensão: {Vsmax:.4f} V")
```

```python

Vin = 15           # Tensão de entrada (V) 24v ~ 30v
Vout = 2         # Tensão de saída (V)
Iout = 2           # Corrente de saída (A)
Pout = Vout * Iout # Potência de saída (W)
fs = 60e3          # Frequência de comutação (Hz)
ripple_IL_percent = 0.05   # Ripple de corrente no indutor (5%)
ripple_Vc_percent = 0.01  # Ripple de tensão no capacitor (1%)

# Cálculo do ciclo de trabalho (Duty Cycle)
D = Vout / Vin
print(f"Ciclo de trabalho (D): {D:.4f}")

# Resistência de carga
Rload = Vout / Iout
print(f"Resistência de carga: {Rload:.2f} Ω")

# Ripple de corrente no indutor em Amperes
delta_IL = ripple_IL_percent * Iout
print(f"Ripple de corrente no indutor (ΔIL): {delta_IL:.4f} A")

# ✔️ Cálculo do indutor
L = (Vin - Vout) * D / (fs * delta_IL)
print(f"Valor do indutor: {L*1e6:.2f} µH")

# ✔️ Cálculo do capacitor
C = delta_IL / (8 * fs * ripple_Vc_percent * Vout)
print(f"Valor do capacitor: {C*1e6:.2f} µF")

# Esforço na chave (MOSFET)
Ismd = D * Iout
Isrms = (D**0.5) * Iout
Ismax = Iout + delta_IL / 2
Vsmax = Vin

print(f"\nEsforço na chave (MOSFET):")
print(f"  Corrente média: {Ismd:.4f} A")
print(f"  Corrente RMS: {Isrms:.4f} A")
print(f"  Corrente máxima: {Ismax:.4f} A")
print(f"  Tensão : {Vsmax:.2f} V")

# Esforço no diodo
Idmd = (1 - D) * Iout
Idrms = ((1 - D) ** 0.5) * Iout
Idmax = Iout + delta_IL / 2

print(f"\nEsforço no diodo:")
print(f"  Corrente média: {Idmd:.4f} A")
print(f"  Corrente RMS: {Idrms:.4f} A")
print(f"  Corrente máxima: {Idmax:.4f} A")
print(f"  Tensão : {Vsmax:.2f} V")
```
